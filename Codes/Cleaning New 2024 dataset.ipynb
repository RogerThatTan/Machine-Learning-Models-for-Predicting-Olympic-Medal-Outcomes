```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import os
from tqdm import tqdm
```

```python
# Load the dataset
df_athlete = pd.read_csv('/content/drive/MyDrive/ML files/2024_New dataset/athletes.csv')
medal_df = pd.read_csv('/content/drive/MyDrive/ML files/2024_New dataset/medallists.csv')
pd.set_option('display.max_rows', None)  # Shows all rows
pd.set_option('display.max_columns', None)  # Shows all columns
```

```python
# prompt: Now merge df_athlete and medal_df as a new table called df where the code column of the df_athlete is equal to code_athlete of the medal_df. Any missing value should be kept empty or NaN.

import pandas as pd
df = pd.merge(df_athlete, medal_df, left_on='code', right_on='code_athlete', how='outer')
```

```python
# prompt: Add three new columns to backup2 after the country_code column called Year, Season, Host_City whose values for every row will be 2024, Summer, Paris respectively.

backup2.insert(backup2.columns.get_loc('country_code') + 1, 'Year', 2024)
backup2.insert(backup2.columns.get_loc('country_code') + 2, 'Season', 'Summer')
backup2.insert(backup2.columns.get_loc('country_code') + 3, 'Host_City', 'Paris')

backup2.head()
```

```python
# prompt: Add a new column called Age after the gender column in backup2 where the value will be a whole integer value which will be the subtraction of birth_date column from the Year column

import pandas as pd
from datetime import datetime

# Convert 'birth_date' to datetime objects
backup2['birth_date'] = pd.to_datetime(backup2['birth_date'], errors='coerce')

# Calculate age based on the year and birth_date
backup2['Age'] = (backup2['Year'] - backup2['birth_date'].dt.year).astype('Int64')

# Reorder columns to place 'Age' after 'gender'
cols = backup2.columns.tolist()
age_index = cols.index('Age')
gender_index = cols.index('gender')
cols.insert(gender_index + 1, cols.pop(age_index))
backup2 = backup2[cols]

backup2.head()
```

```python
# prompt: Remove the third bracket, i.e., [ ] from the disciplines and events columns of backup2

backup2['disciplines'] = backup2['disciplines'].str.replace('[', '', regex=False).str.replace(']', '', regex=False)
backup2['events'] = backup2['events'].str.replace('[', '', regex=False).str.replace(']', '', regex=False)

backup2.head()
```

```python
# prompt: Remove the single quotation i.e.,  ' ' , from the  disciplines column and double quotations , i.e., " ", from the events column of backup2

backup2['disciplines'] = backup2['disciplines'].str.replace("'", "")
backup2['events'] = backup2['events'].str.replace('"', "")

backup2.head()
```

```python
# prompt: Merge the disciplines and events columns as a new column called Events by ensuring a space for backup2 and drop the two of them

# Combine 'disciplines' and 'events' into a new column 'Events'
backup2['Events'] = backup2['disciplines'] + ' ' + backup2['events']

# Drop the original 'disciplines' and 'events' columns
backup2 = backup2.drop(['disciplines', 'events'], axis=1)

backup2.head()
```

```python
# prompt: Drop the birth_rate column

backup2 = backup2.drop('birth_date', axis=1)
backup2.head()
```
