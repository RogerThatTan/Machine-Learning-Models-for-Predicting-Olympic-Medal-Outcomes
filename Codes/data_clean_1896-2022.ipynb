```python
merged_df_new = pd.merge(merged_df, df_uncleaned, on='athlete_id', how='inner')
pd.set_option('display.max_rows', None)  # Shows all rows
pd.set_option('display.max_columns', None)  # Shows all columns
merged_df_new.head()
```

```python
columns_to_drop = [
    'born_city', 'born_region', 'born_country', 'NOC_x', 'team', 'place', 'tied',
    'Roles', 'Full name', 'Used name', 'Born', 'Died', 'NOC_y', 'Measurements',
    'Affiliations', 'Nick/petnames', 'Title(s)', 'Other names', 'Nationality',
    'Original name', 'Name order'
]

merged_df_new.drop(columns=columns_to_drop, inplace=True)
merged_df_new.head()
```

```python
# Convert 'born_date' to datetime
merged_df_new['born_date'] = pd.to_datetime(merged_df_new['born_date'], errors='coerce')

# Calculate age
merged_df_new['age'] = merged_df_new['year'] - merged_df_new['born_date'].dt.year

# Display the first few rows to verify
merged_df_new.head(1)
```

```python
columns_to_drop = [
    'born_date', 'died_date'
]

merged_df_new.drop(columns=columns_to_drop, inplace=True)
merged_df_new.head(1)
```

```python
# Merge 'discipline' and 'event' columns into a single column named 'Event'
merged_df_new['Event'] = merged_df_new['discipline'] + ' ' + merged_df_new['event']

# Drop the original 'discipline' and 'event' columns
merged_df_new.drop(columns=['discipline', 'event'], inplace=True)

# Display the first few rows to verify
merged_df_new.head(1)
```

```python
# Rearrange the columns in the specified order
columns_order = ['athlete_id', 'name', 'Sex', 'age', 'height_cm', 'weight_kg', 'noc', 'year', 'type', 'Event', 'as', 'medal']
merged_df_new = merged_df_new[columns_order]

# Display the first few rows to verify
merged_df_new.head(1)
```

```python
merged_df_new.drop(columns='as', inplace=True)
merged_df_new.head(1)
```

```python
merged_df_new.rename(columns={
    'type': 'Season',
    'Sex':'Gender',
    'weight_kg': 'Weight_KG',
    'height_cm': 'Height_CM',
    'age': 'Age',
    'name': 'Name',
    'noc': 'NOC',
    'year': 'Year',
    'medal': 'Medal'
}, inplace=True)

merged_df_new.head(1)
```

```python
import matplotlib.pyplot as plt
# Group by 'region' and sum the medal counts
medal_counts = backup_2.groupby('NOC')[['Gold', 'Silver', 'Bronze']].sum()

# Sort the medal counts in descending order
medal_counts = medal_counts.sort_values(by=['Gold', 'Silver', 'Bronze'], ascending=False)

# Plot the horizontal bar graph
fig, ax = plt.subplots(figsize=(20, 50))

# Plot each medal type
medal_counts[['Bronze', 'Silver', 'Gold']].plot(kind='barh', stacked=True, color=['#cd7f32', '#c0c0c0', '#ffd700'], ax=ax)

# Set labels and title
ax.set_xlabel('Total Medals')
ax.set_ylabel('Country')
ax.set_title('Total Medals by Country (1896-2022)')

# Display the total medal count on the bars
for i in ax.patches:
    # Find the position for the text
    width = i.get_width()
    if width > 0:  # Only display text for non-zero values
        ax.text(i.get_x() + width / 2, i.get_y() + i.get_height() / 2, str(int(width)),
                fontsize=8, color='black', va='center', ha='center')

# Adjust layout to prevent label overlap
plt.tight_layout()
ax.invert_yaxis()

# Show the plot
plt.show()
```
